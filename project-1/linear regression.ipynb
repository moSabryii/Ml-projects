```python
import pandas as pd

#importing our visulization module
import matplotlib.pyplot as plt
# the as part is a short for our library name and is called alias

#to read the csv file
data = pd.read_csv('cost_revenue_clean.csv')
data

from sklearn.linear_model import LinearRegression
```

```python
#gives you information about your data
data.describe()
```

```python
#to make a datafarame from our csv for a column we use:
x = pd.DataFrame(data,columns=['production_budget_usd'])
y = pd.DataFrame(data,columns=['worldwide_gross_usd'])
y.describe()
```

```python
#to change the figure size
plt.figure(figsize=(10,6))#(width,height)

#to scatter our x and y dataframes
plt.scatter(x,y,alpha=0.3)#alpha defines the opacity

#to title our scatter
plt.title('Movie cost vs Global revenue')

#to name x and y labels
plt.xlabel('production_budget_usd')
plt.ylabel('worldwide_gross_usd')

#to get rid of empty spaces in our chart
plt.ylim(0,3000000000)#(max,min)
plt.xlim(0,450000000)#(max,min)


#to show our visualization
plt.show()
```

Creating object from our model:

```python
regression = LinearRegression()
```

```python
#fitting our model to our data
regression.fit(x,y)
```

```python
#coefficiency slope
regression.coef_
```

```python
#line intercept
regression.intercept_
```

Plotting the model

```python
#to change the figure size
plt.figure(figsize=(10,6))#(width,height)

#to scatter our x and y dataframes
plt.scatter(x,y,alpha=0.3)#alpha defines the opacity

#plotting our prediction line
plt.plot(x,regression.predict(x), color='red', linewidth=4)

#to title our scatter
plt.title('Movie cost vs Global revenue')

#to name x and y labels
plt.xlabel('production_budget_usd')
plt.ylabel('worldwide_gross_usd')

#to get rid of empty spaces in our chart
plt.ylim(0,3000000000)#(max,min)
plt.xlim(0,450000000)#(max,min)


#to show our visualization
plt.show()
```

```python
#evlauation of model
regression.score(x,y)
```
