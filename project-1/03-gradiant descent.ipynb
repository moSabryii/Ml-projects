```python
import matplotlib.pyplot as plt
import numpy as np
```

## Example 1 - A simple cost function

### $$f(x) = x^2 + x +1$$

```python
def f(x):
    return x**2 + x + 1
```

```python
#Make Data
x_1 = np.linspace(start=-3, stop=3, num=500)
x_1
#linspace is a numpy function used to generate sequence of data between 
#start and stop parameters and num parameter specify the
#number of values generated in the array
```

```python
#plot 
plt.plot(x_1,f(x_1))
plt.xlim(-3,3)
plt.ylim(0,8)
plt.xlabel('X')
plt.ylabel('f(x)')
```
